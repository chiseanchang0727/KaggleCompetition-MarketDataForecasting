```python
import os
from dotenv import load_dotenv

import polars as pl
import pandas as pd

import matplotlib.pyplot as plt

load_dotenv(dotenv_path='.env')
```

```python
if load_dotenv(dotenv_path='.env'):
    train_path = os.getenv('TRAIN_PATH')
else: 
    print("")
    # kaggle data path
```

```python
df =pd.read_parquet(os.path.join(train_path, 'partition_by_symbol', 'df_symbol_1.parquet'))
```

```python
# df = df_target
# Plotting
plt.figure(figsize=(20, 6))
plt.plot(df.index, df["responder_6"], linestyle='-', color='black', linewidth=0.1)

# Customizing the plot
plt.title("Symbold_1 in Responder_6")
plt.xlabel("Index")
plt.ylabel("Responder_6")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
plt.plot(df.index, df["responder_6"].cumsum(), linestyle='-', color='black', linewidth=0.1)

# Customizing the plot
plt.title("Symbold_1 in Responder_6")
plt.xlabel("Index")
plt.ylabel("Responder_6")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
